Difference between standard deviation and standard error of the mean: 

The standard deviation measures the spread of the individual data points in a dataset, whereas the standard error of the mean estimates the accuracy with which the sample mean represents the population mean. If the standard deviation is large, it indicates that the data points are spread out over a wide range of values and if it is small, it indicates that the data points are close to the mean. It is used to describe the variability within a single dataset. On the other hand, the standard error of the mean indicates how much the sample mean is expected to vary if we take multiple samples from the population. It is used when you want to investigate how the mean of a sample relates to the population mean.

Here are the steps to create a 95% confidence interval: 
1. calculate the standard error of the mean (SEM) by using this formula: SEM = SD/√n where SD is the standard deviation of the sample and n is the sample size. 
2. Determine the margin of error by multifpying the SEM by the critical value from the standard normal distribution for a 95% confidence level (approximately 1.96). 
3. Create the confidence interval by using this formula: confidence interval = mean ± (1.96×SEM)

````python
import numpy as np

# Sample data (you can replace this with your actual sample)
sample = np.random.normal(loc=50, scale=10, size=20)  # Mean = 50, Std Dev = 10, Sample size = 20

# Function to perform bootstrap and calculate a 95% confidence interval for the mean
def bootstrap_ci(data, num_bootstrap=1000, statistic=np.mean, ci_level=0.95):
    np.random.seed(42)  # For reproducibility
    bootstrapped_stats = []
    
    # Create bootstrap samples and calculate the statistic (mean in this case)
    for _ in range(num_bootstrap):
        resample = np.random.choice(data, size=len(data), replace=True)
        bootstrapped_stats.append(statistic(resample))
    
    # Calculate the percentiles for the confidence interval
    lower_bound = np.percentile(bootstrapped_stats, (1 - ci_level) / 2 * 100)
    upper_bound = np.percentile(bootstrapped_stats, (1 + ci_level) / 2 * 100)
    
    return lower_bound, upper_bound

# Calculate 95% bootstrap confidence interval for the population mean
ci_mean = bootstrap_ci(sample)
print(f"95% Bootstrap Confidence Interval for the mean: {ci_mean}")
````

````python
# Calculate 95% bootstrap confidence interval for the population median
ci_median = bootstrap_ci(sample, statistic=np.median)
print(f"95% Bootstrap Confidence Interval for the median: {ci_median}")
````

````python
# You can also calculate the confidence interval for other statistics like variance, mode, etc.
# Example: Calculate 95% bootstrap confidence interval for the population variance
ci_variance = bootstrap_ci(sample, statistic=np.var)
print(f"95% Bootstrap Confidence Interval for the variance: {ci_variance}")
````

1. Bootstrapping is a way to figure out how accurate a guess or estimate is, without needing lots of math or assumptions. Imagine you want to know the average height of all people in a city, but you only have a small group of people. Instead of just trusting that one group's average, you randomly pick people from that group over and over (sometimes repeating people). Each time, you calculate the average again. Doing this a lot gives you a bunch of new averages, which helps you see how much the estimates vary. This variation gives you a better idea of how close your original guess might be to the true average for the whole city.

2. The main purpose of bootstrapping is to see how reliable your estimates are when you only have a small amount of data. Instead of just guessing and hoping your number is close to the real answer, bootstrapping lets you create many new “fake” samples from your original data. By doing this, you can check how much your estimate (like an average) changes. This helps you understand how confident you can be in your guess, even if you don’t have a lot of data or don’t know much about the whole population. It’s a way to measure how trustworthy your result is. 

3. For each of these new samples, you’d calculate a new average. After doing this a lot, you’d end up with a whole bunch of new averages.Now, you can check if the average you guessed (the hypothesized value) falls in the range of these new bootstrapped averages. If your guess fits nicely within that range, it’s probably a reasonable guess. But if it’s way outside of the range, your guess might not be right.

When a confidence interval overlaps zero, it indicates that there may be no effect at all and that zero could be one possible number for the drug's genuine effect. The confidence interval informs you that there is still a possibility that the true effect (or no effect) could be zero, even though the sample mean you observed is not zero. You "fail to reject the null hypothesis" as a result, since the evidence leaves open the option that the medication is ineffective. On the other hand, if the confidence interval doesn’t include zero (meaning all the possible values for the true effect are either positive or negative), you can “reject the null hypothesis.” This means you're more confident the drug is having an effect because zero is no longer a likely option.

````python
from scipy import stats

# Hypothesis Testing
t_stat, p_value = stats.ttest_rel(df['FinalHealthScore'], df['InitialHealthScore'])
print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
````
